#### Summary
Solar storms often impact the Earth in many different ways including different events on the surface of the Earth (power grid blackouts, damage satellites and can increase atmospheric drag) along with impacts to the magentosphere. We will explore how these impacts we know of on Earth compare to Mars.

#### Background
Data from NASA's MAVEN and ESA's Mars Express (during solar events in 2017 and 2024) have provided evidence that since Mars has little to no magnetic barrier, solar storms often cause an increase of extreme radiation levels on the surface. Often during solar storms on Earth, we are protected by the shape of the magnetic field since it acts like a rock in the river and deflects the plasma and harmful radiation from the surface. 



Exploring how this is different from how Mars' surface responds to solar storms can be helpful in terms of exploring the habitability of the planet and making important differences between the two planets. 

#### Problem Statement 

We will explore how solar storms impact both the weather on Earth and magnetic fields and then take our findings of that and compare it to the "disorganized" magnetic field of Mars. 

- How do the magnetic field characteristics of Mars impact it's habitability?
